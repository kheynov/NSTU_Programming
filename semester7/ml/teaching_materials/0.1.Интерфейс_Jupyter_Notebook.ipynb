## Типы ячеек. Код
Содержит код, который будет быть выполнен. Результат вычислерний последней строки будет выведен под ячейкой

```python
# Комментарий строчный
"""
Комментарий блочный
"""

ThisIsCodeVar = 7  # код программы
```

```python
# Пример вывода вычислений в ячейке.
2 + 3  # Это вычисление не будет выведено
2 ** 3  # т.к. выводится результат последней строки
```

```python
# Пример произвольного вывода вычислений в ячейке.
print(2 + 3)
print(2 ** 3)
```

```python
import time  # импорт библиотеки

# "Долгое вычисление"
for item in range(0, 10000):
    time.sleep(1)
    print(item, end="_")
```

## Команда оценки времени выполнения ячейки
__%%time__ позволяет получить информацию о времени работы кода в рамках одной ячейки.

```python
%%time
import time

time.sleep(1)
```
